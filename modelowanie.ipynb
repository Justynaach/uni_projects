```python
from scipy.stats import norm
import numpy as np
from scipy.stats import binom
```

1. Wagi worków czerwonego żwiru mogą być modelowane rozkładem normalnym o średniej 25,8 kg i odchyleniu standardowym 0,5 kg. 
(a) Wyznacz prawdopodobieństwo, że losowo wybrany worek czerwonego żwiru będzie 
ważył: 

(i) mniej niż 25 kg; 

```python
x=25
srednia=25.8
odch_st=0.5
wynik1=norm.cdf(x, loc=srednia, scale=odch_st)

print(f'Prawdopodobieństwo, że waga będzie niższa niż 25kg to {wynik1:.2f} kg')
```

(ii) pomiędzy 25,5 kg a 26,5 kg.

```python
x1=25.5
x2=26.5

wynik2=norm.cdf(x2, loc=srednia, scale=odch_st)-norm.cdf(x1, loc=srednia, scale=odch_st)
print(f'Prawdopodobieństwo wagi worków pomiędzy 25.5kg a 26.5kg to {wynik2:.2f} kg')
```

(b) Wyznacz, z dokładnością do dwóch miejsc po przecinku, wagę, którą przekracza 75% 
worków.

```python
q=0.25  #lewy ogon
wynik3=norm.ppf(q, loc=srednia, scale=odch_st)
print(f'Waga przekraczająca 75% to {wynik3:.2f} kg')
```

2. Czas, X minut, potrzebny Fredowi na zainstalowanie tachimetru może być traktowany jako 
zmienna losowa o rozkładzie normalnym o średniej 134 i odchyleniu standardowym 16.

```python
odch=16
sr=134
X=150

wynik1=norm.cdf(X, loc=sr, scale=odch)
print(f'Wynik to {wynik1:.2f}')
```

(ii) Wyznacz, z dokładnością do jednego miejsca po przecinku, czas przekraczany przez 10% 
instalacji.

```python
q=0.9
wynik2=norm.ppf(q, loc=sr, scale=odch)

print(f'Czas przekraczany przez 10% instalacji to {wynik2:.1f} min')
```

(b) Czas, Y minut, potrzebny Tadeuszowi na zainstalowanie tachimetru może również być 
zmienną losową o rozkładzie normalnym, lecz wiadomo, że: 
P(Y <170 )=0.14 oraz P(Y>200)=0.03
Wyznacz, do najbliższej minuty, wartości średniej i odchylenia standardowego Y.

```python
z1=norm.ppf(0.14)
z2=norm.ppf(0.97)
odch=(200-170)/(z2-z1)
sr=200-(z2*odch)
###########################################

print(f"Odchylenie standardowe to {odch:.2f}")
print(f"Średnia to {sr:.2f}")
```

3. Objętość L litrów farby emulsyjnej w plastikowym pojemniku może być modelowana rozkładem normalnym o średniej 10.25 i wariancji var.
(a) Zakładając, że var=0.04 wyznacz P(L<10)  

```python
srednia=10.25
var=0.04
sigma=np.sqrt(var)
wynik=norm.cdf(10, loc=srednia, scale=sigma)
print(f'Wynik to {wynik:.2f}')
```

(b) Znajdź wartość odchylenia standardowego tak, aby 98% pojemników zawierało więcej niż 10 litrów farby.

```python
z=norm.ppf(0.02)  #lewy ogon
odchylenie=(10-srednia)/z   #przeksztalcenie wzoru standaryzacji
print(f'Odchylenie standardowe wynosi {odchylenie:.2f}')
```

7. Dostawca gazu utrzymuje zespół inżynierów, którzy są dostępni do usuwania zgłoszonych 
przez klientów wycieków. Większość wycieków można usunąć szybko, lecz niektóre 
wymagają dużo czasu. Czas (bez czasu podróży) potrzebny na usunięcie wycieku ma średnią 
65 minut i odchylenie standardowe 60 minut. 
(a) Zakładając, że czasy można modelować rozkładem normalnym, oszacuj 
prawdopodobieństwo, że: 
(i) usunięcie wycieku zajmie więcej niż 185 minut; 

```python
srednia=65
odch=60
# P(X > 185) = 1 - P(X < 185)
wynik=1-norm.cdf(185, loc=srednia, scale=odch)

print(f'Prawdopodobieństwo P(X > 185) ={wynik:.2f}')
```

(ii) usunięcie wycieku zajmie od 50 do 125 minut; 

```python
wynik =norm.cdf(125, loc=srednia, scale=odch)-norm.cdf(50, loc=srednia, scale=odch)
print(f'Prawdopodobieństwo (50 < X < 125) to {wynik:.2f}')
```

(iii) średni czas usuwania wycieków w losowej próbie 90 wycieków będzie mniejszy niż 70 
minut.

```python
nowe_odch=odch/np.sqrt(90)
wynik=norm.cdf(70, loc=srednia, scale=nowe_odch)

print(f'Prawdopodobieństwo to {wynik:.2f}')
```
